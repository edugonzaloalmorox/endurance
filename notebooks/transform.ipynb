## Clean bikepacking.com info

```python
bikepacking = pd.read_csv('data/df_endurance.csv')
bikepacking.head(10)
```

### Clean names of races

```python

# Make column names lowercase

bikepacking.columns = bikepacking.columns.str.lower()

```

```python
# Make age column numeric

bikepacking['age'] = pd.to_numeric(bikepacking['age'], errors='coerce')
```

```python
bikepacking['race'] = bikepacking['race'].str.replace('the', '')
bikepacking['race'] = bikepacking['race'].str.replace('The', '')
```

```python
bikepacking.head(10)
```

```python
bikepacking['race'].unique()
```

```python
bikepacking['source'] = 'bikepacking.com'
bikepacking.head(10) 
```

```python
skim(bikepacking)
```
